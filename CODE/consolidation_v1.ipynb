# STAY 22


```python
stay22 = stay22[['Product', 'Provider', 
                 'Booked Date', 'Booking Status', 
                 'Nights', 'Campaign IDs','Commission']]
```

```python
stay22 = stay22[stay22['Campaign IDs'].fillna('').str.startswith("petitfute, a")]
```

```python
stay22['article_id'] = stay22['Campaign IDs'].str.extract(r'(a\d+)')
```

```python
stay22
```

```python
# Agrégation des commissions par article_id
commission_sum = stay22.groupby('article_id', as_index=False)['Commission'].sum()
commission_sum = commission_sum.rename(columns={'Commission': 'Total Commission'})

# Left join sur article_id
merged = base_articles.merge(
    commission_sum,
    on='article_id',
    how='left'
)
```
